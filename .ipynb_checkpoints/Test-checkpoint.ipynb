```python
from tensorflow.keras.models import load_model
 
# load model
model = load_model('inceptionV3.h5')
# summarize model.
model.summary()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cv2

import numpy as np
import pandas as pd
import random
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input
```

```python
output_labels = [ 'Other dogs', 'Jack Russell'] 
index = list(range(0, 2))
labels = dict(zip(index, output_labels))
labels
```

```python
img = cv2.imread('test_images/Image_3.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb = cv2.resize(img_rgb, (299, 299))
plt.figure(figsize=(4, 4))
plt.imshow(img_rgb)
plt.axis('off')
test_img = preprocess_input(img_rgb)
test_img = np.expand_dims(img_rgb, axis=0) / 255.0
probs = model.predict(test_img)
predicted_class = labels[np.argmax(probs)]
for idx in probs.argsort()[0]:
    print("{:.2f}%".format(probs[0][idx]*100), "\t", labels[idx])
```
